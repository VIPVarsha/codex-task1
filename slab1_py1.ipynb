Using the Pandas library, load a CSV file and perform basic data analysis tasks, such as calculating the average of a selected column. Additionally, use Matplotlib to create visualizations, including bar charts, scatter plots, and heatmaps, to analyze the data. Provide insights and observations based on the analysis and visualizations.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from sklearn.preprocessing import LabelEncoder

# Uploads the  CSV file
uploaded = files.upload()  # You will be prompted to upload a file

# Reads the CSV file
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# this displays dataset information
print(df.info())
print(df.describe())

# this identifies the categorical columns and encode them
categorical_columns = df.select_dtypes(include=['object']).columns
if len(categorical_columns) > 0:
    print(f"Encoding categorical columns: {list(categorical_columns)}")
    le = LabelEncoder()
    for col in categorical_columns:
        df[col] = le.fit_transform(df[col])

# this chooses a numeric column for analysis
selected_column = "column_name"  # Replace with actual numeric column name
if selected_column in df.columns:
    average_value = df[selected_column].mean()
    print(f"Average of {selected_column}: {average_value}")

# Bar Chart
categorical_column = "category_column"  # Replaces with actual categorical column name
if categorical_column in df.columns:
    df[categorical_column].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Distribution of {categorical_column}")
    plt.xlabel(categorical_column)
    plt.ylabel("Count")
    plt.show()

# Scatter Plot
x_column = "column_x"  # Replaces with actual numeric column
y_column = "column_y"  # Replaces with actual numeric column
if x_column in df.columns and y_column in df.columns:
    plt.scatter(df[x_column], df[y_column], color='red', alpha=0.5)
    plt.title(f"{x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
